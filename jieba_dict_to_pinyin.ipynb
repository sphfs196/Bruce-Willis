```python
# -*- coding: utf-8 -*-
import logging
import jieba
from opencc import OpenCC 
import pandas as pd
from pypinyin import pinyin,slug
```

```python
#用opencc把結疤字典的簡體轉繁體
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
texts_num = 0
openCC = OpenCC('s2tw')
f_in = open("C:/Users/miplab01/Desktop/jupyter notebook files/Bruce-Willis/jieba_dict_tw.txt",'w',encoding='utf-8')

with open("C:/Users/miplab01/Desktop/jupyter notebook files/jieba-master/extra_dict/dict.txt.big",'r',encoding='utf-8') as jieba_dict:
    for line in jieba_dict:
        f_in.write(openCC.convert(line))
        texts_num += 1
        if texts_num % 10000 == 0:
                logging.info("已處理 %d 行" % texts_num)

                
f_in.close()
```

```python
dataSet_A = set()
dataSet_B = set()
dataSet_C = set()
dataSet_D = set()
dataSet_E = set()
dataSet_F = set()
dataSet_G = set()
dataSet_H = set()
dataSet_I = set()
dataSet_J = set()
dataSet_K = set()
dataSet_L = set()
dataSet_M = set()
dataSet_N = set()
dataSet_O = set()
dataSet_P = set()
dataSet_Q = set()
dataSet_R = set()
dataSet_S = set()
dataSet_T = set()
dataSet_U = set()
dataSet_V = set()
dataSet_W = set()
dataSet_X = set()
dataSet_Y = set()
dataSet_Z = set()
dataSet = [dataSet_A,dataSet_B,dataSet_C,dataSet_D,dataSet_E,dataSet_F,dataSet_G,dataSet_H,dataSet_I,dataSet_J,dataSet_K,dataSet_L,dataSet_M,dataSet_N,dataSet_O,dataSet_P,dataSet_Q,dataSet_R,dataSet_S,dataSet_T,dataSet_U,dataSet_V,dataSet_W,dataSet_X,dataSet_Y,dataSet_Z]
```

```python
with open("C:/Users/miplab01/Desktop/jupyter notebook files/Bruce-Willis/jieba_dict_tw.txt",'r',encoding='utf-8') as jieba_dict:
    for texts_num, line in enumerate(jieba_dict):
        line = line.strip('\n')
        word = line.split(" ")
        word.append(slug(word[0],style = 8))
        if word[3][0] == 'a':
            dataSet_A.add(tuple(word))
        elif word[3][0] == 'b':
            dataSet_B.add(tuple(word))
        elif word[3][0] == 'c':
            dataSet_C.add(tuple(word))
        elif word[3][0] == 'd':
            dataSet_D.add(tuple(word))
        elif word[3][0] == 'e':
            dataSet_E.add(tuple(word))
        elif word[3][0] == 'f':
            dataSet_F.add(tuple(word))
        elif word[3][0] == 'g':
            dataSet_G.add(tuple(word))
        elif word[3][0] == 'h':
            dataSet_H.add(tuple(word))
        elif word[3][0] == 'i':
            dataSet_I.add(tuple(word))
        elif word[3][0] == 'j':
            dataSet_J.add(tuple(word))
        elif word[3][0] == 'k':
            dataSet_K.add(tuple(word))
        elif word[3][0] == 'l':
            dataSet_L.add(tuple(word))
        elif word[3][0] == 'm':
            dataSet_M.add(tuple(word))
        elif word[3][0] == 'n':
            dataSet_N.add(tuple(word))
        elif word[3][0] == 'o':
            dataSet_O.add(tuple(word))
        elif word[3][0] == 'p':
            dataSet_P.add(tuple(word))
        elif word[3][0] == 'q':
            dataSet_Q.add(tuple(word))
        elif word[3][0] == 'r':
            dataSet_R.add(tuple(word))
        elif word[3][0] == 's':
            dataSet_S.add(tuple(word))
        elif word[3][0] == 't':
            dataSet_T.add(tuple(word))
        elif word[3][0] == 'u':
            dataSet_U.add(tuple(word))
        elif word[3][0] == 'v':
            dataSet_V.add(tuple(word))
        elif word[3][0] == 'w':
            dataSet_W.add(tuple(word))
        elif word[3][0] == 'x':
            dataSet_X.add(tuple(word))
        elif word[3][0] == 'y':
            dataSet_Y.add(tuple(word))
        elif word[3][0] == 'z':
            dataSet_Z.add(tuple(word))
            
        if (texts_num + 1) % 10000 == 0:
            logging.info("已完成前 %d 行轉換分類" % (texts_num + 1))
```

```python
index = ord('A')
filePath = "C:/Users/miplab01/Desktop/jupyter notebook files/Bruce-Willis/jieba_letter/"
fileName = "jieba_letter_" + chr(index) + ".csv"
data = []
data_pseg = []
data_pinyin = []
data_freq = []

for data_letter in dataSet:
    data = []
    data_pseg = []
    data_pinyin = []
    data_freq = []
    data_letter_lst = list(data_letter)
    #print(data_letter_lst)
    for item in data_letter_lst:
        data.append(item[0])
        data_pseg.append(item[2])
        data_pinyin.append(item[3])
        data_freq.append(item[1])
    output_csv = pd.DataFrame({
        'chinese' : data,
        'posseg' : data_pseg,
        'pinyin' : data_pinyin,
        'frequency' : data_freq
    })
    fileName = "jieba_letter_" + chr(index) + ".csv"
    output_csv.to_csv(filePath + fileName,index = False, encoding='utf-8')
    index += 1

```
